# Creation des data sets à clusteriser

```python
# création d'un DataFrame contenant customer_id, customer_unique_id et city_rank
# city_rank correspond au rang de la ville :
# 5 pour plus de 2 000 000, 4 pour plus de 1 000 000, 3 pour plus de 500 000,
# 2 pour plus de 200 000, 1 pour plus de 100 000 et 0 en dessous 
# (source https://en.wikipedia.org/wiki/List_of_cities_in_Brazil_by_population)

cities = pd.read_csv("villes.csv", header=None, names=["customer_city", "city_rank"])
cities["customer_city"] = cities["customer_city"].map(lambda x: x.lower())
cities["customer_city"] = cities["customer_city"].map(lambda x: x.strip(' '))
customers = raw_customers[["customer_id", "customer_unique_id", "customer_city"]]
customers = pd.merge(customers, cities, on="customer_city", how="left")
customers.drop(columns=["customer_city"], inplace=True)
customers = customers.fillna(0)
customers
```

```python
# création d'un DataFrame contenant order_purchase_timestamp, order_id, customer_id, city_rank
# et customer_unique_id avec les commandes dont le order_status est delivered.
# Ce DataFrame, df_orders, est indéxé par date

df_orders = orders.loc[orders["order_status"]=="delivered",["order_id","customer_id","order_purchase_timestamp"]]
df_orders = pd.merge(df_orders, customers, on="customer_id", how="left")
df_orders = df_orders.set_index("order_purchase_timestamp")
df_orders.head()
```

```python
# déclinaisons de df_orders par périodes 

df_orders_quarter = df_orders.to_period(freq="Q")
df_orders_month = df_orders.to_period(freq="M")
df_orders_week = df_orders.to_period(freq="W")
```

```python
# stockage des DataFrames par période dans des dictionnaires

list_quarters = df_orders_quarter.sort_index(inplace=False).index.unique()
dict_df_orders_quarter = dict()
for index in list_quarters:
    df = df_orders_quarter.loc[index]
    if isinstance(df, pd.core.series.Series):
        df = df.to_frame().transpose()
    dict_df_orders_quarter[str(index)] = df

list_month = df_orders_month.sort_index(inplace=False).index.unique()
dict_df_orders_month = dict()
for index in list_month:
    df = df_orders_month.loc[index]
    if isinstance(df, pd.core.series.Series):
        df = df.to_frame().transpose()
    dict_df_orders_month[str(index)] = df
del dict_df_orders_month["2016-09"]
```

```python
# regroupement des catégories

grouped_categories = {
"body" : ["beleza_saude", "perfumaria", "fraldas_higiene"],
"electronic_computers" : ["informatica_acessorios", "telefonia", "tablets_impressao_imagem", "telefonia_fixa",
                        "consoles_games", "eletronicos", "pcs", "pc_gamer"],
"cars" : ["automotivo"],
"house" : ["cama_mesa_banho", "moveis_decoracao", "utilidades_domesticas", "ferramentas_jardim", "eletroportateis",
         "climatizacao", "moveis_cozinha_area_de_servico_jantar_e_jardim", "moveis_escritorio", "eletrodomesticos", 
         "casa_conforto", "moveis_colchao_e_estofado", "moveis_sala", "moveis_quarto", "eletrodomesticos_2", 
         "la_cuisine", "casa_conforto_2", "portateis_casa_forno_e_cafe", 
         "portateis_cozinha_e_preparadores_de_alimentos"],
"leisure_gifts" : ["esporte_lazer", "relogios_presentes", "papelaria", "cool_stuff", "artigos_de_festas", 
                 "artigos_de_natal", "flores"],
"food" : ["alimentos_bebidas", "market_place", "alimentos", "bebidas"],
"family_pets" : ["bebes", "brinquedos", "pet_shop", "other"],
"clothes_fashion" : ["fashion_bolsas_e_acessorios", "malas_acessorios", "fashion_roupa_masculina", 
                   "fashion_underwear_e_moda_praia", "fashion_esporte", "fashion_roupa_feminina", 
                   "fashion_roupa_infanto_juvenil", "fashion_calcados"],
"art_music_books" : ["audio", "livros_tecnicos", "instrumentos_musicais", "artes", "livros_interesse_geral", 
                   "livros_importados", "cine_foto", "musica", "cds_dvds_musicais", "dvds_blu_ray", 
                   "artes_e_artesanato"],
"professionnal" : ["construcao_ferramentas_construcao", "agro_industria_e_comercio", "construcao_ferramentas_jardim", 
                 "construcao_ferramentas_ferramentas", "casa_construcao", "construcao_ferramentas_iluminacao",
                 "industria_comercio_e_negocios", "construcao_ferramentas_seguranca", "sinalizacao_e_seguranca",
                 "seguros_e_servicos"]
}
```

```python
# fonction de remplissage des features

# liste des moyens de paiement
list_payment_types = list(payments["payment_type"].unique())

# liste des catégories
list_categories = list(products["product_category_name"].dropna().unique())
list_categories.append("other")

def features(data):
    """
    Cette fonction reçoit en paramètre un DataFrame du type df_orders et renvoie
    un DataFrame avec les customer_unique_id en index, complèté avec les features suivantes :
        - nombre de commandes ("orders_count")
        - taux moyen de satisfaction ("review_score_mean")
        - utilisateur ou non des commentaires ("review_user")
        - nombre d'articles commandés ("nb_items")
        - somme totale dépensée ("total spent")
        - somme moyenne dépensée par article ("mean_spent")
        - nombre d'articles commandés par catégorie (cf list_categories)
        - nombre d'utilisation des moyens de paiements (cf list_payment type)
    """
    # on ne garde qu'une valeur de chaque customer_unique_id avec son city_rank
    returned_data = data.groupby("customer_unique_id").city_rank.mean()
    n_obs = returned_data.shape[0]
    
    # on ajoute le nombre de commandes
    df = data.groupby("customer_unique_id").agg(
        orders_count = pd.NamedAgg("order_id", "count"))
    returned_data = pd.merge(returned_data, df, on="customer_unique_id")
    
    # on ajoute le taux moyen de satisfaction et l'usage des commentaires
    df = pd.merge(data, reviews[["order_id", "review_score", "review_comment_title", "review_comment_message"]],
                  on="order_id", how="left")
    df1 = df.groupby("customer_unique_id").agg(
        review_score_mean = pd.NamedAgg("review_score", "mean"))
    returned_data = pd.merge(returned_data, df1, on="customer_unique_id")
    df["review_user"] =  ((~(df["review_comment_title"].isnull())) | (~(df["review_comment_message"].isnull())))
    df1 = df.groupby("customer_unique_id").agg(
        review_user = pd.NamedAgg("review_user", "sum"))
    returned_data = pd.merge(returned_data, df1, on="customer_unique_id")
    
    # on ajoute le nombre d'objets, leurs catégories, la somme totale et la dépense moyenne
    df = pd.merge(data, items[["order_id", "order_item_id", "product_id", "price", "freight_value"]],
                  on="order_id", how="left")
    df = pd.merge(df, products[["product_id", "product_category_name"]], 
                  on="product_id", how="left")
    df["product_category_name"] = df["product_category_name"].fillna("other")
    df1 = df.groupby("customer_unique_id").agg(
        nb_items = pd.NamedAgg("product_id", "count"),
        total_prices = pd.NamedAgg("price", "sum"),
        total_freight = pd.NamedAgg("freight_value", "sum"))
    returned_data = pd.merge(returned_data, df1, on="customer_unique_id")
    returned_data = returned_data.assign(total_spent=returned_data["total_prices"]+returned_data["total_freight"])
    returned_data.drop(columns=["total_prices", "total_freight"], inplace=True)
    returned_data = returned_data.assign(mean_spent=returned_data["total_spent"]/returned_data["nb_items"])
    df1 = df.groupby("customer_unique_id").product_category_name.value_counts()
    df1 = df1.unstack(level=-1, fill_value=0)
    returned_data = pd.merge(returned_data, df1, on="customer_unique_id")
    missing_cols = set(list_categories) - set(df1.columns)
    for cat_name in missing_cols:
        returned_data = returned_data.assign(new_col=np.zeros(n_obs, dtype=int))
        returned_data.rename(columns={"new_col" : cat_name}, inplace=True)
    
    # on regroupe les catégories
    for key_list, list_cat in grouped_categories.items():
        returned_data[key_list] = returned_data[list_cat].sum(axis=1)
        returned_data.drop(columns=list_cat, inplace=True)
    
    # on ajoute les moyens de paiement utilisés
    df = pd.merge(data, payments[["order_id", "payment_type"]], on="order_id", how="left")
    df = df.groupby("customer_unique_id").payment_type.value_counts()
    df = df.unstack(level=-1, fill_value=0)
    returned_data = pd.merge(returned_data, df, on="customer_unique_id")
    missing_cols = set(list_payment_types) - set(df.columns)
    for payment_type in missing_cols:
        returned_data = returned_data.assign(new_col=np.zeros(n_obs, dtype=int))
        returned_data.rename(columns={"new_col" : payment_type}, inplace=True)
     
    return returned_data
```

```python
# remplissage des features pour chaque période et stockage dans des dictionnaires

dict_quarter = dict()
for period, df in dict_df_orders_quarter.items():
    try :
        dict_quarter[period] = features(df)
    except :
        print(f"Période {period} inexploitable")
    
dict_month = dict()
for period, df in dict_df_orders_month.items():
    try :
        dict_month[period] = features(df)
    except :
        print(f"Période {period} inexploitable")
```

```python
# enregistrement des DataFrames utilisables

pickle.dump(dict_quarter, open("pickle_dict_quarter.pkl", "wb"))
pickle.dump(dict_month, open("pickle_dict_month.pkl", "wb"))
```
